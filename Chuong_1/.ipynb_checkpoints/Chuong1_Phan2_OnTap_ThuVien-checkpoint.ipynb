## 1.Ôn tập thư viện Numpy
Là thư viện số, chuyên xử lý dữ liệu dạng mảng nhiều chiều 

```python
import numpy as np

#Tạo 1 ma trận các số ngẫu nhiên trong khoảng từ 1-100

matranA = np.random.randint(1,100, (10,10))
print(matranA)
print(" ")
print("Kích thước ma trận A : ", matranA.shape) #shape cho biết kích thước hàng và cột


```

```python
#Tách ma trận con số 1: Lấy 10 hàng và tất cả cột
matran_A1 = matranA[:,0:10]
matran_A1
```

```python
#Tách ma trận con số 2: Lấy tất cả các hàng và 5 cột cuối
matran_A2 = matranA[:,10:]
print(matran_A2)
print(matran_A2.shape)
```

```python
#Truy cập dữ liệu Numpy

#Hàng đầu
hangdau = matranA[0,:]
hangdau
```

```python
#Hàng cuối
hangcuoi = matranA[-1,:]
hangcuoi
```

```python
#Lấy phần tử cuối
phantucuoi = matranA[-1,-1]
phantucuoi
```

## 2. Ôn tập Matplotlib
Sử dụng để trực quan hóa, vẽ các biểu đồ


```python
import matplotlib.pyplot as plt

ten_khoa_x = ['K64','K65','K66','K67','K68']
so_hs_y = [450, 570, 1520, 820, 200]

#Vẽ biểu đồ
```

```python
#Vẽ biểu đồ dạng cột

plt.bar(ten_khoa_x,so_hs_y) # .bar là biểu đồ dạng cột

plt.title("Biểu đồ sinh viên từng năm") #Tên biểu đồ

plt.xlabel("Khóa học") #Tên trục x

plt.ylabel("Số lượng") #Tên trục y

plt.show()

```

```python
#Vẽ biểu đồ dạng cột nằm ngang

plt.barh(ten_khoa_x,so_hs_y, color='green') # .barh là biểu đồ dạng cột nằm ngang

plt.title("Biểu đồ sinh viên từng năm") #Tên biểu đồ

plt.xlabel("Số lượng") #Tên trục x

plt.ylabel("Khóa học") #Tên trục y

plt.show()

```

```python
#Vẽ biểu đồ dạng đường 
plt.plot(ten_khoa_x, so_hs_y, 'r--*', ms=10) # màu đỏ / nét đứt / chấm * / ms là khoảng cách nét đứt

plt.show()

```

```python
#Vẽ biểu đồ tròn

plt.pie(so_hs_y, labels=ten_khoa_x,
       autopct = '%2f%%',
       wedgeprops = dict(width = 0.7, edgecolor='w')) #labels nhãn của biểu đồ / autopct lấy thông số phần trăm

plt.show()
```

```python
#Hiển thị 2 biểu đồ trong cùng 1 khung
plt.figure(figsize=(14,5))

plt.subplot(1,2,1)
#Vẽ biểu đồ dạng thẳng đứng
plt.bar(ten_khoa_x, so_hs_y, color="brown")
plt.grid(ls="--")
plt.xlabel("Khóa")
plt.ylabel("Số lượng")

plt.subplot(1,2,2)
#Vẽ biểu đồ tròn

plt.pie(so_hs_y, labels=ten_khoa_x,
       autopct = '%2f%%',
       wedgeprops = dict(width = 0.7, edgecolor='w')) #labels nhãn của biểu đồ / autopct lấy thông số phần trăm

plt.show()
```

```python
#Đọc dữ liệu trong file Excel Data_Movies.xlsx

df_movie1 = pd.read_excel("Data\Data_Movies.xlsx",
                        sheet_name='1900s')

df_movie1
```
